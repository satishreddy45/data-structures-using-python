```python
def merge_sort(arr):
    if len(arr) <= 1:
        return  # Base case: a single element array is already sorted

    mid = len(arr) // 2
    left = arr[:mid]  # Split the array into two halves
    right = arr[mid:]

    # Recursively sort both halves
    merge_sort(left)
    merge_sort(right)

    # Merge the sorted halves back into the original array
    merge_two_sorted_list(left, right, arr)

def merge_two_sorted_list(a, b, arr):
    len_a = len(a)
    len_b = len(b)

    i = j = k = 0  # Pointers for a, b, and arr

    # Merge the two sorted arrays
    while i < len_a and j < len_b:
        if a[i] <= b[j]:
            arr[k] = a[i]
            i += 1
        else:
            arr[k] = b[j]
            j += 1
        k += 1

    # Copy any remaining elements from a
    while i < len_a:
        arr[k] = a[i]
        i += 1
        k += 1

    # Copy any remaining elements from b
    while j < len_b:
        arr[k] = b[j]
        j += 1
        k += 1

# Input array
test_cases = [
    [5, 3, 8, 6, 2, 7, 4, 1],
    [1, 2, 3, 4, 5, 6, 7, 8],
    [8, 7, 6, 5, 4, 3, 2, 1],
    [4, 2, 7, 4, 3, 2, 1, 7],
    [10]
]

for arr in test_cases:
    merge_sort(arr)
    print(arr)
```

```python
def merge_sort1(elements, key):
    if len(elements) <= 1:
        return

    # Split the list into two halves
    mid = len(elements) // 2
    left = elements[:mid]
    right = elements[mid:]

    # Recursively sort both halves
    merge_sort1(left, key)
    merge_sort1(right, key)

    # Merge the sorted halves
    merge_two_sorted_list1(left, right, elements, key)

def merge_two_sorted_list1(left, right, elements, key):
    len_left = len(left)
    len_right = len(right)

    i = j = k = 0  # Pointers for left, right, and elements

    # Merge based on the given key
    while i < len_left and j < len_right:
        if left[i][key] <= right[j][key]:
            elements[k] = left[i]
            i += 1
        else:
            elements[k] = right[j]
            j += 1
        k += 1

    # Copy any remaining elements from the left list
    while i < len_left:
        elements[k] = left[i]
        i += 1
        k += 1

    # Copy any remaining elements from the right list
    while j < len_right:
        elements[k] = right[j]
        j += 1
        k += 1

# List of dictionaries
elements = [
    {'name': 'rajab', 'age': 12, 'time_hours': 3},
    {'name': 'vignesh', 'age': 21, 'time_hours': 2.5},
    {'name': 'chinmay', 'age': 24, 'time_hours': 1.5},
    {'name': 'vedanth', 'age': 17, 'time_hours': 1},
]


merge_sort1(elements, 'name')
print("Sorted by name:")
for elem in elements:
    print(f"  Name: {elem['name']}, Age: {elem['age']}, Time Hours: {elem['time_hours']}")

merge_sort1(elements, 'age')
print("\nSorted by age:")
for elem in elements:
    print(f"  Name: {elem['name']}, Age: {elem['age']}, Time Hours: {elem['time_hours']}")

merge_sort1(elements, 'time_hours')
print("\nSorted by time_hours:")
for elem in elements:
    print(f"  Name: {elem['name']}, Age: {elem['age']}, Time Hours: {elem['time_hours']}")
```
